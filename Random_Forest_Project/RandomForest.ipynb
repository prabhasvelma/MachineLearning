```python
import pandas as pd
from sklearn import datasets
```

```python
x=datasets.load_iris()
```

```python
x.keys()
```

```python
df=pd.DataFrame(x.data)
df
```

```python
df['target']=x.target
df
```

```python
x.target_names
```

```python
df['target_name']=df.target.apply(lambda val:x.target_names[val])
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
sample=df.drop('target_name',axis='columns')
sample
```

```python
x_train,x_test,y_train,y_test=train_test_split(sample.drop('target',axis='columns'),sample.target,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=40)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
